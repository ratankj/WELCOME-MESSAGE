## code  for New meter

##### New meter in week: 99 denotes new meter. it will replace by 1 and other values like 1,2,3 replace by 0

```python

WELCOME_MSG_UNIQUE_RECORD_FIRST_WEEK = WELCOME_MSG_UNIQUE_RECORD[WELCOME_MSG_UNIQUE_RECORD['Week']==min(WELCOME_MSG_UNIQUE_RECORD['Week'].unique())]

dataframe = []
dataframe.append(WELCOME_MSG_UNIQUE_RECORD_FIRST_WEEK)

for i in range(2 , max(WELCOME_MSG_UNIQUE_RECORD['Week']) + 1):

    # PREVIOUS WEEK DATA
    previous_week_data = WELCOME_MSG_UNIQUE_RECORD[WELCOME_MSG_UNIQUE_RECORD['Week'] == i-1]
    #previous_week_data_count = previous_week_data['New Meter No'].count()

    # CURRENT WEEK DATA
    current_week_data = WELCOME_MSG_UNIQUE_RECORD[WELCOME_MSG_UNIQUE_RECORD['Week'] == i]
    #current_week_data_count = current_week_data['New Meter No'].count()

    MERGE_CURRENT_PREVIOUS_DATA = pd.merge(current_week_data,previous_week_data[['New Meter No','Week']],left_on=['New Meter No'], right_on=['New Meter No'], how='left')
    
    # after join which are blank fill them with 99
    MERGE_CURRENT_PREVIOUS_DATA['Week_y'] = MERGE_CURRENT_PREVIOUS_DATA['Week_y'].fillna(99)
    MERGE_CURRENT_PREVIOUS_DATA['Week_y'] = MERGE_CURRENT_PREVIOUS_DATA['Week_y'].astype(int)




    dataframe.append(MERGE_CURRENT_PREVIOUS_DATA)

# Concatenate all DataFrames in the list along the rows
MERGE_CURRENT_PREVIOUS_DATA = pd.concat(dataframe, ignore_index=True)

# fill 1 in week_x null rows.  bcz while calculating and appending the first week data. it show null
# fill 99 in the week_y null rows. bcz these are related to week 1. and will replace 99 by 1 . and other value by 0
# 1 indicate new value and 0 indicate old meter

MERGE_CURRENT_PREVIOUS_DATA['Week_x'] = MERGE_CURRENT_PREVIOUS_DATA['Week_x'].fillna(min(MERGE_CURRENT_PREVIOUS_DATA['Week'].unique()))
MERGE_CURRENT_PREVIOUS_DATA['Week_y'] = MERGE_CURRENT_PREVIOUS_DATA['Week_y'].fillna(99)

# data change to int type
MERGE_CURRENT_PREVIOUS_DATA['Week_x'] = MERGE_CURRENT_PREVIOUS_DATA['Week_x'].astype(int)
MERGE_CURRENT_PREVIOUS_DATA['Week_y'] = MERGE_CURRENT_PREVIOUS_DATA['Week_y'].astype(int)


MERGE_CURRENT_PREVIOUS_DATA.drop(columns=['Week'], inplace=True)
MERGE_CURRENT_PREVIOUS_DATA.rename(
    columns = {"Week_x":"Week","Week_y":"NEW_METER_COUNT"},inplace=True)


#MERGE_CURRENT_PREVIOUS_DATA['New_meter_count_per_week'] = MERGE_CURRENT_PREVIOUS_DATA['New_meter_count_per_week'].replace(99, 1).apply(lambda x: 0 if x != 1 else x)
# Output data file 

MERGE_CURRENT_PREVIOUS_DATA.to_csv(r"C:\Users\Ratan Kumar Jha\Desktop\welcome msg\ANALYSIS OUTPUT\MERGE_CURRENT_PREVIOUS_DATA.csv")



```

```python
# after partition or group by we did lead of detailed remark

welcome_message_data['lead_Detailed_remark'] = welcome_message_data.groupby('New Meter No')['Detailed_Remarks'].shift(-1)
```

```python
import pandas as pd

output_list = []

for meter, group in welcome_message_data.groupby('New Meter No'):
    sum_meter_aging = 0  

    for i in range(len(group)):
        starting_remark = group['Detailed_Remarks'].iloc[i]
        lead_remark = group['lead_Detailed_remark'].iloc[i]
        meter_aging_value = group['meter_aging'].iloc[i]

        if starting_remark != lead_remark:
            sum_meter_aging += meter_aging_value  
            date = group['Aging_Creation_Date'].iloc[i]  
            output_list.append({
                'meter_no': meter,
                'date': date,
                'Detailed_Remark': starting_remark,
                'Sum_meter_aging': sum_meter_aging
            })
        
            sum_meter_aging = 0
        else:
          
            sum_meter_aging += meter_aging_value  
           

# Convert to DataFrame
Dataframe_meter_compiled_category = pd.DataFrame(output_list)

# Print the DataFrame
#print(output_df)
Dataframe_meter_compiled_category.to_csv(r"C:\Users\Ratan Kumar Jha\Desktop\welcome msg\ANALYSIS OUTPUT\FINAL_DATA\Dataframe_meter_compiled_category.csv")
```
